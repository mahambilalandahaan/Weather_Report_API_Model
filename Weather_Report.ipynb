```python
## *~*~*~*~*â›ˆWEATHER REPORT API PROJECTâ›ˆ *~*~*~*~*

#IMPORTS
!pip install -q requests
import requests
from datetime import datetime, timedelta
import warnings
from google.colab import userdata
warnings.filterwarnings("ignore")
api_key = userdata.get('OR_API_KEY')
#APPLYING CONDITION
while True:
    print("\n press y for continue \n press n for exit")
    press = input("Do you want to continue (y/n)?")
    if press.lower() == 'y' :
        city = input("Enter City Name  ")

# USING SECRET KEY/TOKEN
        #api_key=os.getenv('OR_API_KEY')

        params = {'q':city,'appid': api_key}
#API MODEL
        response = requests.get('https://api.openweathermap.org/data/2.5/weather',params=params)
        data = response.json()

#FETCHING DATA
        if 'cod' in data and data['cod'] != 200:
            print(f"\n Error: {data['message']}")
            print("Please enter a valid city name.")
        else:

            city = data['name']
            country = data['sys']['country']
            temperature = data['main']['temp']       # Kelvin
            feels_like = data['main']['feels_like']  # Kelvin
            weather = data['weather'][0]['description']
            humidity = data['main']['humidity']
            wind_speed = data['wind']['speed']
            wind_speed_kmh = wind_speed * 3.6
            offset_time=data['timezone']
            utc_time= datetime.utcnow()
            local_time = utc_time + timedelta(seconds=offset_time)
            local_time = local_time.strftime("%Y-%m-%d %H:%M:%S")

            sunrise_utc = datetime.utcfromtimestamp(data['sys']['sunrise'])
            sunset_utc = datetime.utcfromtimestamp(data['sys']['sunset'])

            sunrise_local = sunrise_utc + timedelta(seconds=offset_time)
            sunset_local = sunset_utc + timedelta(seconds=offset_time)

#MARKDOWN TABLE
            # Markdown table with symbols in content
            markdown_table = f"""
            ### ğŸŒ¤ï¸ Weather Report for {city}, {country}

            | **Field**        | **Value** |
            |:-----------------|-----------------------------:|
            | City       ğŸŒ†   | {city}                  |
            | Country    ğŸŒ   | {country}               |
            | Temperature ğŸŒ¡   | {temperature - 273.15:.2f}Â°C/ {(temperature - 273.15) *(9/5)+32:.2f} F       |
            | Feels Like   ğŸŒ¡  | {feels_like - 273.15:.2f}Â°C / {(feels_like - 273.15)*(9/5)+32:.2f}F       |
            | Weather          | `{weather}`               |
            | Humidity    â„   | {humidity}%               |
            | Wind Speed   â›±  | {wind_speed} m/s  ({wind_speed_kmh:.2f} km/h)       |
            | Sunrise     ğŸŒ…  | {sunrise_local.strftime('%Y-%m-%d %H:%M:%S')}           |
            | Sunset     ğŸŒ‡   | {sunset_local.strftime('%Y-%m-%d %H:%M:%S')}            |
            | Local Time  ğŸ••     | {local_time}              |

              """

            # Print table to console
            print(markdown_table)

            # Save to Markdown file
            with open("Weather_Report.md", "a") as f:
                f.write(markdown_table + "\n\n")
                f.write(f"\n\n Report generated at ğŸ• {local_time}")

            print("âœ… Markdown report updated  as latest Weather_Report.md")

    elif press.lower() == 'n':
        break
    else:
     print("Invalid input. Please enter 'y' or 'n'.")
```
